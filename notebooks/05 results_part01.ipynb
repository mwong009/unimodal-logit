```python
# data to generate tables for paper

import os
import pickle
import pandas as pd

with open("../models/m1_ordinallogit.pickle", "rb") as f:            
    m1_ordinallogit = pickle.load(f)

with open("../models/m2_unimodallogit.pickle", "rb") as f:            
    m2_unimodallogit = pickle.load(f)

with open("../models/m3_zerotrunc_unimodallogit.pickle", "rb") as f:            
    m3_zerotrunc_unimodallogit = pickle.load(f)
```

```python
print('m1 ll', m1_ordinallogit.logLike)
print('m1 bic', m1_ordinallogit.bayesian)
print('m1 $\\bar{\\rho}^2$', m1_ordinallogit.rhoSquareNull)
print('m1 time', m1_ordinallogit.optimizationMessages['Optimization time'])

print('m2 ll', m2_unimodallogit.logLike)
print('m2 bic', m2_unimodallogit.bayesian)
print('m2 $\\bar{\\rho}^2$', m2_unimodallogit.rhoSquareNull)
print('m2 time', m2_unimodallogit.optimizationMessages['Optimization time'])

print('m3 ll', m3_zerotrunc_unimodallogit.logLike)
print('m3 bic', m3_zerotrunc_unimodallogit.bayesian)
print('m3 $\\bar{\\rho}^2$', m3_zerotrunc_unimodallogit.rhoSquareNull)
print('m3 time', m3_zerotrunc_unimodallogit.optimizationMessages['Optimization time'])
# m1_ordinallogit.__dict__
```
